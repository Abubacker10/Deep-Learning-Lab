```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread('/content/Screenshot 2024-07-03 101255.png')

g_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

ret, thresh1 = cv2.threshold(g_image, 127, 255, cv2.THRESH_BINARY)
ret, thresh2 = cv2.threshold(g_image, 127, 255, cv2.THRESH_BINARY_INV)
ret, thresh3 = cv2.threshold(g_image, 127, 255, cv2.THRESH_TRUNC)
ret, thresh4 = cv2.threshold(g_image, 127, 255, cv2.THRESH_TOZERO)
ret, thresh5 = cv2.threshold(g_image, 127, 255, cv2.THRESH_TOZERO_INV)

plt.imshow(thresh1)
plt.show()
plt.imshow(thresh2)
plt.show()
plt.imshow(thresh3)
plt.show()
plt.imshow(thresh4)
plt.show()
plt.imshow(thresh5)
plt.show()

if cv2.waitKey(0) & 0xFF == 27:
    cv2.destroyAllWindows()
```
